ðŸ”´Which cities have the highest and lowest air quality?

```python
highest_AQ_city = df.loc[df['AirQuality'].idxmax(), 'City']
lowest_AQ_city = df.loc[df['AirQuality'].idxmin(), 'City']
print("City with the highest AirQuality:", highest_AQ_city)
print("City with the lowest AirQuality:", lowest_AQ_city)
```

ðŸ”´Which cities have the highest and lowest water pollution levels?

```python
highest_pollution_city = df.loc[df['WaterPollution'].idxmax(), 'City']
lowest_pollution_city = df.loc[df['WaterPollution'].idxmin(), 'City']
print("City with the highest water pollution level:", highest_pollution_city)
print("City with the lowest water pollution level:", lowest_pollution_city)
```

ðŸ”´Is there any relationship between air quality and water pollution in the dataset?

```python
correlation = df['AirQuality'].corr(df['WaterPollution'])
print("Correlation between air quality and water pollution:", correlation)
plt.figure(figsize=(8, 6))
sns.scatterplot(x='AirQuality', y='WaterPollution', data=df)
plt.xlabel('AirQuality')
plt.ylabel('WaterPollution')
plt.title('Relationship between Air Quality and Water Pollution')
plt.show()
```
